```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

-----
Uniform Smoothing Filtering
-----

```python
image_in = cv2.imread("/workspaces/LTU-Programming-Portfolio/d0040e_vt2026/session_ip/Assignment 2 Image Processing/Arvid Wallin-4.png")
image_color = cv2.cvtColor(image_in, cv2.COLOR_BGR2RGB)
image_gray = cv2.cvtColor(image_color, cv2.COLOR_BGR2GRAY)
```

```python
kernel3 = np.ones((3,3), np.float32)/9
image_smooth3 = cv2.filter2D(image_color, -1, kernel3)

kernel5 = np.ones((5,5), np.float32)/25
image_smooth5 = cv2.filter2D(image_color, -1, kernel5)

kernel9 = np.ones((9,9), np.float32)/81
image_smooth9 = cv2.filter2D(image_color, -1, kernel9)
```

-----
Display
-----

```python
plt.figure(figsize=(18,24))

plt.subplot(2,2,1)
plt.imshow(image_color)
plt.title("Original Image")
plt.axis("off")

plt.subplot(2,2,2)
plt.imshow(image_smooth3)
plt.title("3x3 smoothing")
plt.axis("off")

plt.subplot(2,2,3)
plt.imshow(image_smooth5)
plt.title("5x5 smoothing")
plt.axis("off")

plt.subplot(2,2,4)
plt.imshow(image_smooth9)
plt.title("9x9 smoothing")
plt.axis("off")

plt.show()
```

-----
Other Filters
-----

```python
box_blur = cv2.boxFilter(image_color, -1, ksize=(9,9), normalize=True)
gaussian_blur = cv2.GaussianBlur(image_color, ksize=(25,25), sigmaX=3)
median_blur = cv2.medianBlur(image_color, ksize=9)
bilaterial_blur = cv2.bilateralFilter(image_color, 9, 25, 25)
```

```python
plt.subplot(2,2,1)
plt.imshow(box_blur)
plt.title("Box Blur")
plt.axis("off")

plt.subplot(2,2,2)
plt.imshow(gaussian_blur)
plt.title("Gaussian Blur")
plt.axis("off")

plt.subplot(2,2,3)
plt.imshow(median_blur)
plt.title("Median Blur")
plt.axis("off")

plt.subplot(2,2,4)
plt.imshow(bilaterial_blur)
plt.title("Bilateral Blur")
plt.axis("off")

plt.show()
```
