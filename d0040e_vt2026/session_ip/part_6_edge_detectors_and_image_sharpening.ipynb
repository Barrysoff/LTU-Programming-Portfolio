```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image_in = cv2.imread("/workspaces/LTU-Programming-Portfolio/d0040e_vt2026/session_ip/Assignment 2 Image Processing/BL_Royal_Vincent_of_Beauvais.jpg")
image_color = cv2.cvtColor(image_in, cv2.COLOR_BGR2RGB)
image_gray = cv2.cvtColor(image_color, cv2.COLOR_BGR2GRAY)
```

-----
Apply Sobel Operator
-----

```python
sobelx = cv2.Sobel(image_gray, cv2.CV_64F, 1, 0, ksize=3)
sobely = cv2.Sobel(image_gray, cv2.CV_64F, 0, 1, ksize=3)
```

-----
Compute Gradient magnitude
-----

```python
gradient_magnitude = cv2.magnitude(sobelx, sobely)
```

-----
Convert to uint8
-----

```python
gradient_magnitude = cv2.convertScaleAbs(gradient_magnitude)
```

-----
Display
-----

```python
#plt.subplot(2,2,1)
plt.imshow(gradient_magnitude)
plt.title("Original Image")
plt.axis("off")
```
