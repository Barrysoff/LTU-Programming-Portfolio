```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

-----
Invert Colors
-----


```python
image_in = cv2.imread("/workspaces/LTU-Programming-Portfolio/d0040e_vt2026/session_ip/Assignment 2 Image Processing/flowers_color.png")
image_color = cv2.cvtColor(image_in, cv2.COLOR_BGR2RGB)
image_gray = cv2.cvtColor(image_color, cv2.COLOR_BGR2GRAY)
```

-----
Apply negative image
-----

```python
image_negative = 255 - image_gray
```

-----
Display
-----

```python
plt.figure(figsize=(18,12))

plt.subplot(1,2,1)
plt.imshow(image_gray, cmap="gray", vmin=0, vmax=255)
plt.title("Original image")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(image_negative, cmap="gray", vmin=0, vmax=255)
plt.title("Inverted image")
plt.axis("off")

plt.show()
```

-----
Adjust Brightness
-----

```python
image_lighter = np.where(image_gray<=255-50, image_gray + 50, 255)
image_darker = np.where(image_gray>50, image_gray - 50, 0)
```

-----
Display
-----

```python
plt.figure(figsize=(18,12))

plt.subplot(1,3,1)
plt.imshow(image_lighter, cmap="gray", vmin=0, vmax=255)
plt.title("Lighter image")
plt.axis("off")

plt.subplot(1,3,2)
plt.imshow(image_gray, cmap="gray", vmin=0, vmax=255)
plt.title("Original image")
plt.axis("off")

plt.subplot(1,3,3)
plt.imshow(image_darker, cmap="gray", vmin=0, vmax=255)
plt.title("Darker image")
plt.axis("off")

plt.show()
```
