```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

-----
Reduce number of gray levels
-----

```python
image_color = cv2.imread("/workspaces/LTU-Programming-Portfolio/d0040e_vt2026/session_ip/Assignment 2 Image Processing/flowers_color.png")
image_gray = cv2.cvtColor(image_color, cv2.COLOR_BGR2GRAY)

image_quantized = (image_gray // 64) * 64
```

-----
Display
-----

```python
plt.figure(figsize=(6,6))
plt.imshow(image_quantized, cmap="gray")
plt.title("Grayscale image, 4 shades")
plt.axis("off")
plt.show()
```

-----
Convert to Binary using a threshold
-----

```python
threshold_value = 127
_, image_bilevel = cv2.threshold(image_gray, threshold_value, 255, cv2.THRESH_BINARY)
```

-----
Display
-----

```python
plt.figure(figsize=(6,6))
plt.imshow(image_bilevel, cmap="gray")
plt.title(f'Threshold image, ${threshold_value}$')
plt.axis("off")
plt.show()
```
