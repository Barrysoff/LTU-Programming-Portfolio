```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from PIL import Image
import numpy as np
```

```python
image = np.asarray(Image.open("/workspaces/LTU-Programming-Portfolio/d0040e_vt2026/Session_5/squirrel.jpg"))
# print(image.shape)

X = image.reshape(-1, 3)
# print(image.shape)
```

```python
kmeans = KMeans(n_clusters=15, n_init=10).fit(X)
segmented_img = kmeans.cluster_centers_[kmeans.labels_]
segmented_img = segmented_img.reshape(image.shape)
```

```python
plt.imshow(segmented_img / 255)
plt.show()
```
